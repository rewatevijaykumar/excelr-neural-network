# Assignment-16-Neural Networks-Forest Fires

```python
# import librabries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
#!pip install keras tensorflow
import keras
from keras.models import Sequential 
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
#import dataset
df = pd.read_csv('dataset/forestfires.csv')
df.head()
```

```python
#delete month and day columns
df.drop(['month','day'],axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['size_category'] = LabelEncoder().fit_transform(df['size_category'])
df.sample(10)
```

```python
#split features and target columns
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Feature Scaling to optimize 
X = StandardScaler().fit_transform(X)
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#Initialize ANN with sequence of layers
def build_classifier(optimizer):
    # Classifier Model
    classifier = Sequential()
    #Add input layer and first hidden layer
    classifier.add(Dense(units = 18, kernel_initializer = 'uniform', activation = 'relu', input_dim=28))
    # Add the hidden layers
    classifier.add(Dense(units = 12, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    # Adding the output layer
    # Probability for the outcome 
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    # Compiling the ANN
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
#build classifier
classifier = KerasClassifier(build_fn = build_classifier)
#parameters
param_grid = {'batch_size':[1,5,10],
             'epochs':[50,100,120],
             'optimizer':['adam','rmsprop']}
#gridsearch
grid = GridSearchCV(estimator = classifier, 
                   param_grid = param_grid, 
                   scoring='accuracy',
                   cv=10,
                   n_jobs=-1,
                   verbose=3)
grid_result = grid.fit(x_train, y_train)
```

```python
best_parameters = grid_result.best_params_
best_accuracy = grid_result.best_score_
```

```python
print('best parameters: {}'.format(best_parameters))
print('best accuracy: {}'.format(best_accuracy))
```

```python
# build neural network with best parameters
classifier = Sequential()
#Add input layer and first hidden layer
classifier.add(Dense(units = 18, kernel_initializer = 'uniform', activation = 'relu', input_dim=28))
# Add the hidden layers
classifier.add(Dense(units = 12, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the output layer
# Probability for the outcome 
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the data
classifier.fit(x_train, y_train, batch_size = 5, epochs = 120)
```

```python
# saving|loading model
from keras.models import load_model
#save trained model in h5 format
classifier.save('forest_fires_model.h5')
#load trained model
#classifier = load_model('forest_fires_model.h5')
```

```python
# make predictions
y_pred = classifier.predict(x_test)
y_pred = [ 1 if y>=0.5 else 0 for y in y_pred ]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d) *100
accuracy = round(np.trace(cm)/np.sum(cm) * 100, 2)
print('Model Accuracy: {}%'.format(accuracy))
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
